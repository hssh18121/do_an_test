```python
import os
import cv2
from patchify import patchify
import matplotlib.pyplot as plt
import numpy as np
import tifffile as tiff
from PIL import Image
```

```python
mask_directory = "./global-land-cover-mapping-openearthmap/label_converted"
rgb_directory = "./global-land-cover-mapping-openearthmap/images"
output_directory = "./crop_output"
```

```python
patch_size = 256
```

```python
for path,subdirs,files in os.walk(rgb_directory):
  dirname = path.split(os.path.sep)[-1]
  if(dirname != os.path.split(rgb_directory)[-1] and dirname != 'test'):
    print(dirname)
    for filename in files:
      print(os.path.join(path, filename))
      image = cv2.imread(os.path.join(path, filename), 1)

      SIZE_X = (image.shape[1]//patch_size)*patch_size #Nearest size divisible by our patch size
      SIZE_Y = (image.shape[0]//patch_size)*patch_size #Nearest size divisible by our patch size
      image = Image.fromarray(image)
      print(image.size)
      image = image.crop((0 ,0, SIZE_X, SIZE_Y))  #Crop from top left corner
      # plt.imshow(image)
      # plt.show()
      image = np.array(image)

      patches_img = patchify(image, (256, 256, 3), step=256)  #Step=256 for 256 patches means no overlap

      number_of_image = 0
      for i in range(patches_img.shape[0]):
          for j in range(patches_img.shape[1]):
              single_patch_img = patches_img[i,j,:,:]
              single_patch_img = single_patch_img[0] #Drop the extra unecessary dimension that patchify adds.
              number_of_image += 1
              remove_tif_from_filename = filename.replace('.tif', '')
              output_path = os.path.join(output_directory, f"{dirname}/{remove_tif_from_filename}_rgb")
              os.makedirs(output_path, exist_ok=True)
              patch_path = os.path.join(output_path, f"{number_of_image}.tif")
              tiff.imwrite(patch_path, single_patch_img)

      print(number_of_image)
```
