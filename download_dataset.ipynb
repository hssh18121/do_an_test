```python
# Import necessary libraries
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Path to the image file
image_path = 'global-land-cover-mapping-openearthmap/images/train/aachen_3.tif'

# Open the image using rasterio
with rasterio.open(image_path) as src:
    # Get width and height of the image
    width = src.width
    height = src.height
    
    # Print the width and height
    print(f'Width: {width} pixels')
    print(f'Height: {height} pixels')
    
    # Read the image
    image = src.read()
    
    # Display the image using matplotlib
    plt.figure(figsize=(10, 10))
    show(image, transform=src.transform)
    # plt.title('Aachen 1')
    plt.show()
```

```python
# Path to the image file
image_path = 'global-land-cover-mapping-openearthmap/label/train/aachen_2.tif'

# Open the image using rasterio
with rasterio.open(image_path) as src:
    # Get width and height of the image
    width = src.width
    height = src.height
    
    # Print the width and height
    print(f'Width: {width} pixels')
    print(f'Height: {height} pixels')
    
    # Read the image
    image = src.read(1)  # Read the first band
    
    # Display the original image using matplotlib
    plt.figure(figsize=(10, 10))
    show(image, transform=src.transform, title="Original Image")
    plt.show()

    # Get all unique pixel values
    unique_values = np.unique(image)
    
    # Print the unique pixel values
    print('Unique pixel values:', unique_values)
    
    # Create a new array where only pixels with value 3 are retained
    modified_image = np.where(image == 4, 4, 0)
    
    # Display the modified image using matplotlib
    plt.figure(figsize=(10, 10))
    show(modified_image, transform=src.transform, title="Modified Image with Pixel Value 4 Retained")
    plt.show()

    # Get all unique pixel values in the modified image
    modified_unique_values = np.unique(modified_image)
    
    # Print the unique pixel values in the modified image
    print('Unique pixel values in modified image:', modified_unique_values)
```
